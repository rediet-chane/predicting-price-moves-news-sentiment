```python
# Technical Indicators Analysis
# Author: Rediet Chane Belay

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("=" * 60)
print("TECHNICAL INDICATORS ANALYSIS")
print("=" * 60)

# Note: You need stock price data from yfinance
# For now, this shows the structure

print("\n1. DATA PREPARATION")
print("-" * 40)
print("Loading stock price data for major tickers: AAPL, MSFT, GOOGL, AMZN, TSLA")
print("Columns: Date, Open, High, Low, Close, Adj Close, Volume")

print("\n2. MOVING AVERAGES")
print("-" * 40)
print("Simple Moving Average (SMA):")
print("  - SMA 20: Short-term trend (≈1 month)")
print("  - SMA 50: Medium-term trend (≈2.5 months)")
print("  - SMA 200: Long-term trend (≈10 months)")
print("\nExponential Moving Average (EMA):")
print("  - More weight to recent prices")
print("  - Reacts faster to price changes")

print("\n3. RELATIVE STRENGTH INDEX (RSI)")
print("-" * 40)
print("RSI measures momentum on scale 0-100")
print("  - RSI > 70: Overbought (potential sell signal)")
print("  - RSI < 30: Oversold (potential buy signal)")
print("  - RSI = 50: Neutral")

print("\n4. MACD (Moving Average Convergence Divergence)")
print("-" * 40)
print("Components:")
print("  - MACD Line: (12-day EMA - 26-day EMA)")
print("  - Signal Line: 9-day EMA of MACD")
print("  - Histogram: MACD Line - Signal Line")
print("\nSignals:")
print("  - MACD crosses above signal: Bullish")
print("  - MACD crosses below signal: Bearish")

print("\n5. VISUALIZATIONS")
print("-" * 40)

# Create sample technical indicator plots
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

# Plot 1: Price with Moving Averages
dates = pd.date_range('2024-01-01', '2024-12-31', freq='D')
prices = 100 + np.cumsum(np.random.randn(len(dates)) * 2)
sma20 = pd.Series(prices).rolling(20).mean()
sma50 = pd.Series(prices).rolling(50).mean()

axes[0].plot(dates, prices, 'b-', label='Close Price', linewidth=1)
axes[0].plot(dates, sma20, 'r-', label='SMA 20', linewidth=1)
axes[0].plot(dates, sma50, 'g-', label='SMA 50', linewidth=1)
axes[0].set_title('Stock Price with Moving Averages')
axes[0].set_ylabel('Price ($)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot 2: RSI
rsi = 50 + 30 * np.sin(np.linspace(0, 8*np.pi, len(dates)))
axes[1].plot(dates, rsi, 'purple', linewidth=1)
axes[1].axhline(y=70, color='r', linestyle='--', alpha=0.5)
axes[1].axhline(y=30, color='g', linestyle='--', alpha=0.5)
axes[1].set_title('Relative Strength Index (RSI)')
axes[1].set_ylabel('RSI')
axes[1].set_ylim(0, 100)
axes[1].grid(True, alpha=0.3)

# Plot 3: MACD
macd = 2 * np.sin(np.linspace(0, 6*np.pi, len(dates)))
signal = 1.5 * np.sin(np.linspace(0, 6*np.pi, len(dates)) + 0.5)
histogram = macd - signal
axes[2].plot(dates, macd, 'b-', label='MACD', linewidth=1)
axes[2].plot(dates, signal, 'r-', label='Signal', linewidth=1)
axes[2].bar(dates, histogram, color='gray', alpha=0.5, label='Histogram')
axes[2].set_title('MACD')
axes[2].set_ylabel('Value')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('technical_indicators.png')
print("Saved: technical_indicators.png")

print("\n" + "=" * 60)
print("TECHNICAL INDICATORS COMPLETE")
print("=" * 60)
```
