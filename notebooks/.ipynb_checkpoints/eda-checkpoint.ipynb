````python

---

## Step 8: Create the EDA Notebook

Create `notebooks/eda.ipynb`:

```python
# Predicting Price Moves with News Sentiment - EDA
# Author: Rediet Chane Belay

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

print("=" * 60)
print("PREDICTING PRICE MOVES WITH NEWS SENTIMENT")
print("Exploratory Data Analysis")
print("=" * 60)

# 1. LOAD DATA (mock data for structure - replace with actual data)
print("\n1. DATA LOADING")
print("-" * 40)

# Note: You need to download the actual dataset from the challenge
# For now, this shows the analysis structure
print("This notebook will analyze financial news headlines and stock prices")

# 2. TEXT LENGTH ANALYSIS
print("\n2. TEXT LENGTH ANALYSIS")
print("-" * 40)
print("Analyzing headline character counts and word counts")
print("- Headline length distribution will show typical news article title lengths")
print("- Short headlines (under 50 chars): Quick updates, earnings releases")
print("- Long headlines (over 100 chars): Detailed analysis, feature stories")

# 3. PUBLISHER ANALYSIS
print("\n3. PUBLISHER ANALYSIS")
print("-" * 40)
print("Identifying most active news sources")
print("Top publishers typically include: Benzinga, MarketWatch, Reuters, Bloomberg")
print("Each publisher may have different writing styles and sentiment tendencies")

# 4. TIME SERIES OF NEWS VOLUME
print("\n4. TIME SERIES OF NEWS VOLUME")
print("-" * 40)
print("Analyzing publication frequency over time")
print("- News volume spikes often coincide with major market events")
print("- Earnings season (Jan, Apr, Jul, Oct) typically shows higher volume")
print("- Fed announcements, economic reports, and geopolitical events")

# 5. KEYWORD AND TOPIC ANALYSIS
print("\n5. KEYWORD AND TOPIC ANALYSIS")
print("-" * 40)
print("Common keywords in financial headlines:")
common_keywords = ['earnings', 'stock', 'price', 'target', 'upgrade', 
                   'downgrade', 'buy', 'sell', 'hold', 'analyst']
for kw in common_keywords:
    print(f"  - {kw}")

print("\nTopic categories identified:")
print("  - Earnings announcements")
print("  - Price target changes")
print("  - Analyst ratings (upgrade/downgrade)")
print("  - Merger and acquisition news")
print("  - Product launches")
print("  - Regulatory news")

# 6. VISUALIZATIONS
print("\n6. GENERATING VISUALIZATIONS")
print("-" * 40)

# Create sample plots (will be replaced with real data)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Headline length distribution
axes[0, 0].hist(np.random.normal(80, 20, 1000), bins=30, color='steelblue', edgecolor='black')
axes[0, 0].set_title('Headline Length Distribution')
axes[0, 0].set_xlabel('Character Count')
axes[0, 0].set_ylabel('Frequency')

# Plot 2: Top publishers
publishers = ['Benzinga', 'MarketWatch', 'Reuters', 'Bloomberg', 'Seeking Alpha']
counts = [3500, 2800, 2200, 1800, 1500]
axes[0, 1].bar(publishers, counts, color='coral')
axes[0, 1].set_title('Top Publishers by Article Count')
axes[0, 1].set_xlabel('Publisher')
axes[0, 1].set_ylabel('Number of Articles')
axes[0, 1].tick_params(axis='x', rotation=45)

# Plot 3: News volume over time
dates = pd.date_range('2024-01-01', '2024-12-31', freq='D')
volume = np.random.poisson(20, len(dates))
axes[1, 0].plot(dates, volume, 'b-', linewidth=0.5)
axes[1, 0].set_title('News Volume Over Time')
axes[1, 0].set_xlabel('Date')
axes[1, 0].set_ylabel('Number of Articles')

# Plot 4: Publication hour distribution
hours = list(range(24))
hourly_counts = np.random.poisson(5, 24)
axes[1, 1].bar(hours, hourly_counts, color='green')
axes[1, 1].set_title('Articles by Hour of Day')
axes[1, 1].set_xlabel('Hour (UTC-4)')
axes[1, 1].set_ylabel('Number of Articles')

plt.tight_layout()
plt.savefig('eda_visualizations.png')
print("Saved: eda_visualizations.png")

print("\n" + "=" * 60)
print("EDA COMPLETE")
print("=" * 60)
````
